```python
def outlier_test(data,variable ):
    q1 = data[variable].quantile(0.25)
    q3 = data[variable].quantile(0.75)
    IQR = q3 - q1
    upper_lim = q3 + (1.5 * IQR)
    lower_lim = q1 - (1.5 * IQR)
    
    
    print(f"{variable}: \t IQR - {IQR}, Upper_limit- {upper_lim}, lower_limit- {lower_lim}")
    return lower_lim, upper_lim
```

```python
# for variable in data_num:
# for variable in data_num.columns:
#     outlier_test(data_num, variable)
```

```python
def replace_with_threshold(data, column_name):
    for variable in column_name:
        low_limit, up_limit = outlier_test(data, variable)
        
#         That two lines are used to replace the outlier by thier lower and upper limit
        data.loc[data[variable]<low_limit, variable]= low_limit
        data.loc[data[variable]>up_limit, variable]= up_limit
           
```
